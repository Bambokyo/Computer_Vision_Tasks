Write a Python function that takes in two grayscale images and a set of coordinates as
inputs, and places the smaller one on top of the larger one at the given set of
coordinates.

```python
def Place_Smaller_on_Top(img1,img2,cors):
    
    # converting both to gray scale if dimentions more than 2
    if len(img1.shape) > 2:
        img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    if len(img2.shape) > 2:
        img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    
    # checking sizes
    if (img1.shape < img2.shape):
        large_img = img2
        small_img = img1
    elif img1.shape > img2.shape or img1.shape == img2.shape:
        large_img = img1
        small_img = img2
    else:
        print("Sizes of images not okay for placing on top of reach other")
        return 
    
    
    h, w = small_img.shape
    large_img[cors[1]:cors[1] + h, cors[0]:cors[0] + w] = small_img
    
    plt.imshow(large_img)
    
```

Write a new function (or modify your old one) to again place a small image into a larger
one at a given set of coordinates, but this time blend the subset image with the
background image by adding the two pixel values together and dividing by two in the
regions of overlap. Submit an example of this function in action.

```python
def Place_Smaller_on_Top_Blended(img1,img2,cors):
    
    # converting both to gray scale if dimentions more than 2
    if len(img1.shape) > 2:
        img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    if len(img2.shape) > 2:
        img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    
    # checking sizes
    if (img1.shape < img2.shape):
        large_img = img2
        small_img = img1
    elif img1.shape > img2.shape or img1.shape == img2.shape:
        large_img = img1
        small_img = img2
    else:
        print("Sizes of images not okay for placing on top of reach other")
        return 

    h, w = small_img.shape
    
    # Calculate overlapping region
    overlap_region = large_img[cors[1]:cors[1] + h, cors[0]:cors[0] + w]
    
    # Blend the images
    blended_region = (small_img + overlap_region) // 2
    
    large_img[cors[1]:cors[1] + h, cors[0]:cors[0] + w] = blended_region
    
    plt.imshow(large_img)
    
```
